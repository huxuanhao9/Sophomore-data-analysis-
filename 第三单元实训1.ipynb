```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
text = np.load("D:/Desktop/data.npy",allow_pickle = True)
test = np.load("D:/Desktop/feature_names.npy",allow_pickle = True)
#分析
#总人口不断增加，
#城镇人口增加，乡村人口减少，城镇增幅大于乡村减幅
#男性人口增幅中期略大于女性，后期相齐平
#预测
#总人口继续增加，城镇人口增加，男女人口差额逐渐稳定
```

```python
data = text[:20,:]
data = data[::-1]
print(data)
name = test[:]
print(name)
```

```python
p=plt.figure(figsize=(15,15))
ax1=p.add_subplot(4,1,1)
plt.scatter(data[:,0],data[:,1],marker='o',c='r')
plt.scatter(data[:,0],data[:,2],marker='o',c='g')
plt.scatter(data[:,0],data[:,3],marker='o',c='b')
plt.scatter(data[:,0],data[:,4],marker='o',c='y')
plt.scatter(data[:,0],data[:,5],marker='o',c='m')
plt.xlabel('时间')
plt.ylabel('人口(万人)')
plt.legend(['年末总人口(万人)','男性人口(万人)','女性人口(万人)','城镇人口(万人)','乡村人口(万人)' ])
plt.title('1996-2015年人口数据特征间的关系(总图)')
plt.savefig("D:/Desktop/1996-2015年人口数据特征间的关系(总图).png")
plt.show()
```
