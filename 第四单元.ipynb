```python
month_1 = [i.month for i in order3['ListingInfo1']]
month_2 = [i.month for i in order3['UserupdateInfo2']]
month_3 = [i.month for i in order1['Listinginfo1']]
month_4 = [i.month for i in order1['LogInfo3']]
print('用户信息更新表中ListingInfo1的月份数据：',month_1[:5])
print('用户信息更新表中UserupdateInfo2的月份数据：',month_2[:5])
print('登录信息表中ListingInfo1的月份数据：',month_3[:5])
print('登录信息表中LogInfo3的月份数据：',month_4[:5])
```

```python
minute1 = pd.PeriodIndex(order3['ListingInfo1'],freq='T')
minute2 = pd.PeriodIndex(order3['UserupdateInfo2'],freq='T')
minute3 = pd.PeriodIndex(order1['Listinginfo1'],freq='T')
minute4 = pd.PeriodIndex(order1['LogInfo3'],freq='T')
timedelta_T1 = minute1 - minute2
timedelta_T2 = minute3 - minute4
print('用户信息更新表中的时间差为：\n',timedelta_T1)
print('登录信息表中的时间差为：\n',timedelta_T2)
```

# <center><size font=6.5>实训3

```python
#掌握分组聚合的原理与步骤，掌握agg和apply聚合方法，掌握transform聚合方法
#分析用户信息更新表和登录信息表时，除了提取时间本身的信息外，还可以结合用户编号进行分组聚合，然后进行组内分析。通过组内分析可以得出每组组
#内的最早和最晚信息更新时间和登陆时间，登录次数和信息。
#用groupby方法对用户信息更新表和登录信息表进行分组
order1Group=order1[['Idx','LogInfo1','LogInfo2','Listinginfo1','LogInfo3']].groupby(by = 'Idx')
print("分组后的order1表为:",order1Group)
order2Group=order2[['Idx','UserInfo_1','UserInfo_2','UserInfo_3','UserInfo_4','WeblogInfo_1']].groupby(by = 'Idx')
print("分组后的order2表为:",order2Group)
order3Group=order3[['Idx','ListingInfo1','UserupdateInfo1','UserupdateInfo2']].groupby(by = 'Idx')
print("分组后的order3表为:",order3Group)
```

```python
#用agg方法求取分组后的最早和最晚更新及登录时间
print(order1[['LogInfo3']].agg(['min','max']))
print(order1[['Listinginfo1']].agg(['min','max']))
print(order3[['ListingInfo1']].agg(['min','max']))
print(order3[['UserupdateInfo2']].agg(['min','max']))
#第一次最早和最晚登录时间分别是2011-01-01，2014-11-10
#第二次最早和最晚登录时间分别是2013-11-01，2014-11-10
#第一次最早和最晚更新时间分别是2013-11-01，2014-11-10
#第二次最早和最晚更新时间分别是2012-09-01，2014-11-04
```

```python
#用size方法求取分组后的数据的信息更新次数与登录次数
print(order1[['LogInfo3']].agg(['size']))
print(order1[['Listinginfo1']].agg(['size']))
print(order3[['ListingInfo1']].agg(['size']))
print(order3[['UserupdateInfo2']].agg(['size']))
#第一次登录次数580551
#第二次登录次数580551
#第一次更新次数372463
#第二次更新次数372463
#虽然次数都没有变化，但是规模逐渐扩大，已经达到十万级别
```
